Required libraries.

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
import pandas as pd
```

Cleaned dataset.

```python
DATASET_PATH = './dataset/weighted-full_data.csv'
df = pd.read_csv(DATASET_PATH)
df
```

```python
df.info()
```

Definition of independent variables.

```python
features = df.loc[:, df.columns != 'Disease']
features
```

Definition of dependent variable.

```python
target = df['Disease']
target
```

Splitting the dataset into training, testing, and validation data.

```python
train_ratio = 0.60
validation_ratio = 0.20
test_ratio = 0.20

X_train, X_test, Y_train, Y_test = train_test_split(features, target, test_size=(1 - train_ratio), random_state=1)
X_train, X_val, Y_train, Y_val = train_test_split(X_train, Y_train, test_size=(test_ratio/(test_ratio + validation_ratio)), random_state=1)
```

### Tabnet

```python
X_train = X_train.to_numpy()
Y_train = Y_train.to_numpy()
X_test = X_test.to_numpy()
Y_test = Y_test.to_numpy()
```

```python
import pytorch_tabnet
from pytorch_tabnet.tab_model import TabNetClassifier
import torch
```

```python
clf1_nopreproc = TabNetClassifier(optimizer_fn=torch.optim.Adam,
                                  optimizer_params=dict(lr=2e-2),
                                  scheduler_params={"step_size": 10,  # how to use learning rate scheduler
                                                    "gamma": 0.9},
                                  scheduler_fn=torch.optim.lr_scheduler.StepLR,
                                  mask_type='entmax'  # "sparsemax"
                                  )
```

```python
clf1_nopreproc.fit(
    X_train, Y_train,
    eval_set=[(X_train, Y_train), (X_test, Y_test)],
    eval_name=['train', 'valid'],
    eval_metric=['accuracy'],
    max_epochs=1000, patience=50,
    batch_size=256, virtual_batch_size=128,
    num_workers=0,
    weights=1,
    drop_last=False
)
```
