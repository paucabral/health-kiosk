Importing the necessary libraries.

```python
import csv
import numpy as np
import pandas as pd
```

Loading the narrowed data.

```python
NARROWED_DATA = 'dataset/narrowed_data.csv'
narrowed_data = pd.read_csv(NARROWED_DATA)
narrowed_data
```

Create a list containing the names of diseases.

```python
narrowed_data_diseases = list(narrowed_data['Disease'])
narrowed_data_diseases
```

Loading the cleaned data set with weights.

```python
CLEANED_DATA = 'dataset/dataset_clean.csv'
cleaned_data = pd.read_csv(CLEANED_DATA)
cleaned_data
```

Limit the dataframe to the instance of narrowed diseases list.

```python
cleaned_data = cleaned_data[cleaned_data['Disease'].isin(narrowed_data_diseases)]
cleaned_data = cleaned_data[['Disease', 'Occurence']]
cleaned_data
```

Eliminate duplicate rows.

```python
cleaned_data = cleaned_data.drop_duplicates()
cleaned_data
```

Merge the data frames based on disease name.

```python
weighted_df = pd.merge(left=narrowed_data, right=cleaned_data, how='left', left_on='Disease', right_on='Disease')
weighted_df
```

Repeat the rows based on the column weight.

```python
weighted_df = weighted_df.loc[weighted_df.index.repeat(weighted_df['Occurence'])]
weighted_df = weighted_df.loc[:, weighted_df.columns != 'Occurence']
weighted_df = weighted_df.reset_index(drop=True)
weighted_df
```

Shuffle the entries.

```python
weighted_df = weighted_df.reindex(np.random.permutation(weighted_df.index))
weighted_df
```

Export the weighted dataset.

```python
CSV_FILEPATH = 'dataset/weighted-narrowed_data.csv'
weighted_df.to_csv(CSV_FILEPATH, index=False)
```
