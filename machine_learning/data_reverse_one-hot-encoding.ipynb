Importing the necessary libraries.

```python
import pandas as pd
import numpy as np
```

Loading the one-hot encoded dataset.

```python
DATASET_PATH = 'dataset/processed_data.csv'
df = pd.read_csv(DATASET_PATH)
df
```

Creating an empty symptoms column.

```python
df['Symptoms'] = ''
df
```

Iterating through each column of each row to append the column names to the symptoms column for those that have a value of `1`.

```python
for col_name in df.columns:
    df.loc[df[col_name] == 1, 'Symptoms'] = df['Symptoms']+'|'+col_name

df
```

Reconstruct the data frame with the `Disease` and the `Symptoms` columns as the only columns.

```python
df_reversed = df[['Disease', 'Symptoms']]
df_reversed
```

Export the dataset.

```python
OUTPUT_DATASET_PATH = 'dataset/reversed_one-hot-encoding_dataset.csv'
df_reversed.to_csv(OUTPUT_DATASET_PATH, index=False)
```
