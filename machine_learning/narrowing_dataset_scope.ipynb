### Symptoms Based

List of short listed symptoms.

```python
# Pain related
symptoms_list = ["ache",
                 "excruciating pain",
                 "headache",
                 "low back pain",
                 "pain",
                 "pain chest",
                 "pain abdominal",
                 "pain back",
                 "pain chest",
                 "pain foot",
                 "pain in lower limb",
                 "pain neck",
                 "painful swallowing",
                 "pleuritic pain",
                 "rest pain",
                 "shooting pain",
                 "side pain"]
```

Find presence of any of the symptoms among the diseases.

```python
mod_symp_df = data[(data[symptoms_list] > 0).any(axis=1)]
mod_symp_df
```

Check number of diseases which meets the minimum number of symptom counts from the symptoms list.

```python
n_symptoms = 2
sum_df = mod_symp_df[symptoms_list].sum(axis=1) > n_symptoms

count = 0
for i in sum_df:
  if i == True:
    count += 1

print("Number of diseases with more than {} symptoms found on the symptoms list: {}".format(n_symptoms, count))
```

Separate the data frame into disease and symptoms.

```python
symp_df1 = mod_symp_df["Disease"]
symp_df2 = mod_symp_df[symptoms_list]
```

Add a column counting the number of symptoms observed.

```python
symp_df2['num_of_symptoms'] = symp_df2[symptoms_list].sum(axis=1)
symp_df2
```

Reconstruct the data frame.

```python
mod_symp_df = mod_symp_df.join(symp_df2['num_of_symptoms'])
mod_symp_df
```

```python
n_symptoms = 2
mod_symp_df = mod_symp_df[mod_symp_df["num_of_symptoms"] > n_symptoms]
mod_symp_df = mod_symp_df[mod_symp_df.columns[:-1]]
mod_symp_df
```
