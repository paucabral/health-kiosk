Importing the necessary libraries.

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import linear_kernel
```

Loading the dataset.

```python
DATASET_FILEPATH = 'dataset/processed_data.csv'
df = pd.read_csv(DATASET_FILEPATH)
df
```

Define indices.

```python
indices = pd.Series(df.index, index=df['Disease'])
indices
```

Placing the symptom values on a matrix of its own.

```python
symp_df = df.loc[:, df.columns != 'Disease']
symp_df
```

Initialize cosine similarities kernel.

```python
cosine_similarities = linear_kernel(symp_df, symp_df)
cosine_similarities
```

```python
disease = df['Disease']

indices = pd.Series(df.index, index=df['Disease'])
indices
```

```python
def disease_predict(original_title):
  idx = indices[original_title]

  sim_scores = list(enumerate(cosine_similarities[idx]))

  sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

  sim_scores = sim_scores[1:31]

  movie_indices = [i[0] for i in sim_scores]

  return disease.iloc[movie_indices]
```
