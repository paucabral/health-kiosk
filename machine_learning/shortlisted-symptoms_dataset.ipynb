Import the necessary libraries.

```python
import csv
import numpy as np
import pandas as pd
```

Load the cleaned and processed dataset.

```python
CSV_FILEPATH = 'dataset/processed_data.csv'
data = pd.read_csv(CSV_FILEPATH)
data
```

### Symptoms Based

List of short listed symptoms.

```python
# Pain related
symptoms_list = ["ache",
                 "excruciating pain",
                 "headache",
                 "low back pain",
                 "pain",
                 "pain chest",
                 "pain abdominal",
                 "pain back",
                 "pain chest",
                 "pain foot",
                 "pain in lower limb",
                 "pain neck",
                 "painful swallowing",
                 "pleuritic pain",
                 "rest pain",
                 "shooting pain",
                 "side pain"]
```

Find presence of any of the symptoms among the diseases.

```python
mod_symp_df = data[(data[symptoms_list] > 0).any(axis=1)]
mod_symp_df
```

Check number of diseases which meets the minimum number of symptom counts from the symptoms list.

```python
n_symptoms = 2
sum_df = mod_symp_df[symptoms_list].sum(axis=1) > n_symptoms

count = 0
for i in sum_df:
  if i == True:
    count += 1

print("Number of diseases with more than {} symptoms found on the symptoms list: {}".format(n_symptoms, count))
```
