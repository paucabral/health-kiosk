Importing the necessary libraries.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import xgboost as xgb
import csv
```

Loading the dataset.

```python
DATASET_PATH = 'dataset/generated_dataset.csv'
df = pd.read_csv(DATASET_PATH)
df
```

Divide the data into train and test sets.

```python
target = df['Disease']
features = df.loc[:, df.columns != 'Disease']


train_ratio = 0.60
validation_ratio = 0.20
test_ratio = 0.20


X_train, X_val, Y_train, Y_val = train_test_split(features, target, test_size=(1 - train_ratio), random_state=1)
X_train, X_test, Y_train, Y_test = train_test_split(X_train, Y_train, test_size=(test_ratio/(test_ratio + validation_ratio)), random_state=1)
```

Define the model.

```python
# Init classifier
xgb_cl = xgb.XGBClassifier()
# Fit
xgb_cl.fit(X_train, Y_train)
```

Checking the accuracy of the model.

```python
# Predict
preds = xgb_cl.predict(X_test)
preds_proba = xgb_cl.predict_proba(X_test)

print(preds)
print(preds_proba)

for i, j in zip(preds, preds_proba):
  print("Prediction: {} | Probability: {}".format(i, np.max(j)))

# Score
model_accuracy = accuracy_score(Y_test, preds)
print('Model Accuracy: {}%'.format(model_accuracy*100))
```
