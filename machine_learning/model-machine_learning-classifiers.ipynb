Required libraries.

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
import pandas as pd
```

Cleaned dataset.

```python
DATASET_PATH = './dataset/generated_dataset.csv'
df = pd.read_csv(DATASET_PATH)
df
```

```python
df.info()
```

Definition of independent variables.

```python
features = df.loc[:, df.columns != 'Disease']
features
```

Definition of dependent variable.

```python
target = df['Disease']
target
```

Splitting the dataset into training, testing, and validation data.

```python
train_ratio = 0.60
validation_ratio = 0.20
test_ratio = 0.20

X_train, X_val, Y_train, Y_val = train_test_split(features, target, test_size=(1 - train_ratio), random_state=1)
X_train, X_test, Y_train, Y_test = train_test_split(X_train, Y_train, test_size=(test_ratio/(test_ratio + validation_ratio)), random_state=1)
```

### Machine Learning Models

#### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dt_clf = DecisionTreeClassifier()
dt_clf.fit(X_train, Y_train)
```

#### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr_clf = LogisticRegression()
lr_clf.fit(X_train, Y_train)
```

#### K-Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
knn_clf = KNeighborsClassifier(n_neighbors=5)
knn_clf.fit(X_train, Y_train)
```

#### Gaussian Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
gnb_clf = GaussianNB()
gnb_clf.fit(X_train, Y_train)
```

#### Multinomial Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB
mnb_clf = MultinomialNB()
mnb_clf.fit(X_train, Y_train)
```

#### Support Vector Machines

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
svm_clf = make_pipeline(StandardScaler(), SVC(gamma='auto'))
svm_clf.fit(X_train, Y_train)
```

#### Predictions evaluation.

```python
dt_pred_train = dt_clf.predict(X_val)
lr_pred_train = lr_clf.predict(X_val)
knn_pred_train = knn_clf.predict(X_val)
gnb_pred_train = gnb_clf.predict(X_val)
mnb_pred_train = mnb_clf.predict(X_val)
svm_pred_train = svm_clf.predict(X_val)
```

```python
dt_pred_test = dt_clf.predict(X_test)
lr_pred_test = lr_clf.predict(X_test)
knn_pred_test = knn_clf.predict(X_test)
gnb_pred_test = gnb_clf.predict(X_test)
mnb_pred_test = mnb_clf.predict(X_test)
svm_pred_test = svm_clf.predict(X_test)
```

Training accuracy of each model.

```python
model_scores = {
    "Decision Tree Classifier": metrics.accuracy_score(Y_val, dt_pred_train)*100,
    "Logistic Regression Classifier": metrics.accuracy_score(Y_val, lr_pred_train)*100,
    "K-Nearest Neighbors Classifier": metrics.accuracy_score(Y_val, knn_pred_train)*100,
    "Gaussian Naive Bayes Classifier": metrics.accuracy_score(Y_val, gnb_pred_train)*100,
    "Multinomial Naive Bayes Classifier": metrics.accuracy_score(Y_val, mnb_pred_train)*100,
    "Support Vector Machines Classifier": metrics.accuracy_score(Y_val, svm_pred_train)*100
}

print("TRAINING ACCURACY:")
for model, model_score in model_scores.items():
    print("{} Model accuracy(in %): {}".format(model, model_score))
    
print("\nWINNING MODEL: ", max(model_scores, key=model_scores.get))
```

Testing accuracy of each model.

```python
model_scores = {
    "Decision Tree Classifier": metrics.accuracy_score(Y_test, dt_pred_test)*100,
    "Logistic Regression Classifier": metrics.accuracy_score(Y_test, lr_pred_test)*100,
    "K-Nearest Neighbors Classifier": metrics.accuracy_score(Y_test, knn_pred_test)*100,
    "Gaussian Naive Bayes Classifier": metrics.accuracy_score(Y_test, gnb_pred_test)*100,
    "Multinomial Naive Bayes Classifier": metrics.accuracy_score(Y_test, mnb_pred_test)*100,
    "Support Vector Machines Classifier": metrics.accuracy_score(Y_test, svm_pred_test)*100
}

print("TESTING ACCURACY:")
for model, model_score in model_scores.items():
    print("{} Model accuracy(in %): {}".format(model, model_score))

print("\nWINNING MODEL: ", max(model_scores, key=model_scores.get))
```
