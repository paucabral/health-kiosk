Required libraries.

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
import pandas as pd
```

Cleaned dataset.

```python
DATASET_PATH = './dataset/weighted-narrowed_data.csv'
df = pd.read_csv(DATASET_PATH)
df
```

```python
df.info()
```

Definition of independent variables.

```python
features = df.loc[:, df.columns != 'Disease']
features
```

Definition of dependent variable.

```python
target = df['Disease']
target
```

Splitting the dataset into training and testing data.

```python
X_train, X_test, Y_train, Y_test = train_test_split(features, target, random_state=1, test_size=0.3)
```

### Machine Learning Models

#### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dt_clf = DecisionTreeClassifier()
dt_clf.fit(X_train, Y_train)
```

#### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr_clf = LogisticRegression()
lr_clf.fit(X_train, Y_train)
```

#### K-Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
knn_clf = KNeighborsClassifier(n_neighbors=5)
knn_clf.fit(X_train, Y_train)
```

#### Gaussian Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
gnb_clf = GaussianNB()
gnb_clf.fit(X_train, Y_train)
```

#### Multinomial Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB
mnb_clf = MultinomialNB()
mnb_clf.fit(X_train, Y_train)
```

#### Support Vector Machines

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
svm_clf = make_pipeline(StandardScaler(), SVC(gamma='auto'))
svm_clf.fit(X_train, Y_train)
```

#### Predictions evaluation.

```python
dt_pred = dt_clf.predict(X_test)
lr_pred = lr_clf.predict(X_test)
knn_pred = knn_clf.predict(X_test)
gnb_pred = gnb_clf.predict(X_test)
mnb_pred = mnb_clf.predict(X_test)
svm_pred = svm_clf.predict(X_test)
```

Accuracy of each model.

```python
model_scores = {
    "Decision Tree Classifier": metrics.accuracy_score(Y_test, dt_pred)*100,
    "Logistic Regression Classifier": metrics.accuracy_score(Y_test, lr_pred)*100,
    "K-Nearest Neighbors Classifier": metrics.accuracy_score(Y_test, knn_pred)*100,
    "Gaussian Naive Bayes Classifier": metrics.accuracy_score(Y_test, gnb_pred)*100,
    "Multinomial Naive Bayes Classifier": metrics.accuracy_score(Y_test, mnb_pred)*100,
    "Support Vector Machines Classifier": metrics.accuracy_score(Y_test, svm_pred)*100
}

print("MODEL ACCURACY:")
for model, model_score in model_scores.items():
    print("{} Model accuracy(in %): {}".format(model, model_score))
    
print("\nWINNING MODEL: ", max(model_scores, key=model_scores.get))
```
